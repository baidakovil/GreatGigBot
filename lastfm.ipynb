```python
from urllib.request import urlopen
from datetime import datetime, timedelta

lastfmuser='Anastasia20009'
daysdelta = 2
hoursdelta = 0
minlistens = 3

pagenum = 0
artistlist = dict()
overtime = False
timedelay = timedelta(days=daysdelta,hours=hoursdelta)
today = datetime.utcnow()

while overtime == False :
    pagenum += 1
    artisturl = 'https://www.last.fm/user/'+lastfmuser+'/library?page='+str(pagenum)
    htmlbyte = urlopen(artisturl).read()
    filename = lastfmuser+'_page_'+str(pagenum)+'.html'
    filehandle = open(filename,'w', encoding="utf-8")
    htmltext = htmlbyte.decode()
    filehandle.write(htmltext)
    filehandle.close()

    trackfound = False
    artistfound = False
    timefound = False

    filehandle = open(filename, encoding="utf-8")

    for line in filehandle :
        if timefound == True :
            if 'span title' not in line :
                continue
            timestr = line.split('"')[1]
            print(timestr)
            timeformatted = datetime.strptime(timestr,"%A %d %b %Y, %I:%M%p")
            artistfound = False
            timefound = False
            if (today - timeformatted) < timedelay :
                artistlist[artist] = artistlist.get(artist,0) + 1 #сделать форматирование по регистру АБВ->Абв
            else :
                print('Overtime')
                overtime = True
                filehandle.close()
                break
        if artistfound == True :
            if 'chartlist-timestamp--lang-en' not in line :
                continue
            timefound = True
        if trackfound == True :
            if 'title=' not in line :
                continue
            artist = line.split('"')[1]                
            trackfound = False
            artistfound = True
            print(artist)
        if 'class="chartlist-artist' not in line :
            continue
        trackfound = True        
filehandle.close()
print (artistlist)
print ('Artist so:',sum(artistlist.values()))

#удалить ненужные списки

provedartistlist = list()
for actartist, listens in artistlist.items() :
    if listens >= minlistens :
        provedartistlist.append(actartist)
print ('provedartistlist:', provedartistlist)
print ('Artist proved:',len(provedartistlist))
```

```python
import urllib.parse

eventfiles = dict()

for eventartist in provedartistlist :
    lastfmeventurl = 'https://www.last.fm/music/'+urllib.parse.quote(eventartist)+'/+events'
    print('Download '+eventartist+' event page...',end='')
    htmlbyte = urlopen(lastfmeventurl).read()
    filename = eventartist+'_lastfmevent.html'
    eventfiles[eventartist]=filename
    filehandle = open(filename,'w', encoding="utf-8")
    htmltext = htmlbyte.decode()
    filehandle.write(htmltext)
    filehandle.close()   
    print('Done.')
print (eventfiles)
```
