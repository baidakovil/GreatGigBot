```python
import pandas as pd
fileHandle = open('worldcities.csv', 'r', encoding="utf-8")
cityDf = pd.read_csv(filepath_or_buffer=fileHandle, sep=',', lineterminator='\n')
cityDf.insert(0,'city_orig',cityDf['city'])
cityDf.insert(7,'admin_name_orig',cityDf['admin_name'])
cityDf
cols = cityDf.columns.tolist()
```

```python
cityDf = cityDf.iloc[:,[0,1,2,5,7,9]]
col = cityDf.columns.tolist()
col
['country','admin_name_orig', 'admin_name','city_orig','city', 'city_ascii']
```

```python
cityDf = cityDf[['country','city', 'city_ascii']]
cityDf
```

```python
from tqdm import tqdm_notebook
def writeFile(country, data):
    dataFilename = 'worldcities/'+country+'_cities.csv'
    fileHandle = open(dataFilename, 'w', encoding="utf-8")
    data.to_csv(path_or_buf=fileHandle, sep='\t', line_terminator='\n')
    
for country in tqdm_notebook(cityDf['country'].unique()):
    data = cityDf.loc[cityDf['country'] == country,['city', 'city_ascii']].reset_index(drop=True)    
    writeFile(country, data)
```

```python
listOfCountries = [country for country in cityDf['country'].unique()]
countriesDf = pd.DataFrame({'country':sorted(listOfCountries)}) 
countriesDf.insert(1,'country_0',['' for i in range(0,237)])
countriesDf.insert(2,'country_1',['' for i in range(0,237)])
```

```python
pd.set_option('display.max_rows', None)
countriesDf
```

```python
countriesDf.loc[countriesDf['country']=='United Kingdom','country_0']='UK'
countriesDf.loc[countriesDf['country']=='Russia','country_0']='Russian Federation'
countriesDf.loc[countriesDf['country']=='Russia','country_1']='Россия'
countriesDf
```

```python
dataFilename = 'worldcities/WorldCountryList.csv'
fileHandle = open(dataFilename, 'w', encoding="utf-8")
countriesDf.to_csv(path_or_buf=fileHandle, sep='\t', line_terminator='\n',quoting=None)
```
